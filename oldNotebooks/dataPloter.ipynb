```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("vgg16ModelHistory.pkl", "rb") as file:
    history = pickle.load(file)
```

```python
print(history.keys())
```

```python

# Retrieve accuracy and validation accuracy
train_accuracy = history['accuracy']
true_positives = history['true_positives']
true_negatives = history['true_negatives']
false_positives = history['false_positives']

# Retrieve the number of epochs
epochs = range(1, len(train_accuracy) + 1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(epochs, true_positives, label='true_positives')
plt.plot(epochs, true_negatives, label='true_negatives')
plt.plot(epochs, false_positives, label='false_positives')

# Adding titles and labels
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Display the plot
plt.show()
```
